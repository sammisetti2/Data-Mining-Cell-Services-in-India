```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

address = "../Data/Sampled and Cleaned data/Sampled_Cleaned_July18.csv"
data1=pd.read_csv(address)

address = "../Data/Sampled and Cleaned data/Sampled_Cleaned_July19.csv"
data2=pd.read_csv(address)

address = "../Data/Sampled and Cleaned data/Sampled_Cleaned_May18.csv"
data3=pd.read_csv(address)

address = "../Data/Sampled and Cleaned data/Sampled_Cleaned_May19.csv"
data4=pd.read_csv(address)

address = "../Data/Sampled and Cleaned data/Sampled_Cleaned_March18.csv"
data5=pd.read_csv(address)

address = "../Data/Sampled and Cleaned data/Sampled_Cleaned_March19.csv"
data6=pd.read_csv(address)

df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
df3=pd.DataFrame(data3)
df4=pd.DataFrame(data4)
df5=pd.DataFrame(data5)
df6=pd.DataFrame(data6)
```

```python
df=pd.DataFrame()
df=df.append(df1,ignore_index=True)
df=df.append(df2,ignore_index=True)
df=df.append(df3,ignore_index=True)
df=df.append(df4,ignore_index=True)
df=df.append(df5,ignore_index=True)
df=df.append(df6,ignore_index=True)
df.to_csv("TotalData_Cleaned_Sampled.csv", index=False) # Merging all 6 months data to a single dataframe
```

```python
df['signal_strength']=-df['signal_strength'] #Replacing negative values with positive values to avoid confusion in visualization
df.to_csv("TotalData_Cleaned_Sampled.csv", index=False)
```

```python
#min-max normalization
df['signal_strength']=((df['signal_strength']-df['signal_strength'].min())/(df['signal_strength'].max()-df['signal_strength'].min()))*(100)
df['speed']=((df['speed']-df['speed'].min())/(df['speed'].max()-df['speed'].min()))*100
df.to_csv("TotalData_Cleaned_Sampled.csv", index=False)
```

```python
#Aggregating idea with vodafone
df['service_provider']=df['service_provider'].replace("IDEA","VODAFONE")
df.to_csv("TotalData_Cleaned_Sampled.csv", index=False)
```
